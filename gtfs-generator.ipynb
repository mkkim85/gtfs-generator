```python
import pandas as pd
import numpy as np
```

```python
def timedelta_to_hhmmss(x):
    hh = str(x.components.hours + (24 * x.components.days)).zfill(2)
    mm = str(x.components.minutes).zfill(2)
    ss = str(x.components.seconds).zfill(2)
    return hh + ':' + mm + ':' + ss
```

```python
R = pd.read_csv('노선.csv')
```

```python
T = pd.read_csv('정류소출도착.csv')
```

```python
service_id = 'mon-tues-wed-thurs-fri-sat-sun'

t_columns = ['route_id', 'service_id', 'trip_id', 'shape_id']
s_columns = ['trip_id', 'arrival_time', 'departure_time', 'stop_id', 'stop_sequence']

df_trips = pd.DataFrame(columns=t_columns)
df_stop_times = pd.DataFrame(columns=s_columns)

df_trips.to_csv('trips.txt', index=False, encoding='utf-8-sig')
df_stop_times.to_csv('stop_times.txt', index=False, encoding='utf-8-sig')

for index, r in R.iterrows():
    df_trips = pd.DataFrame(columns=t_columns)
    df_stop_times = pd.DataFrame(columns=s_columns)
    
    route_id = r['\ufeffISC노선ID']
    t = T.loc[T['\ufeffISC노선ID'] == route_id]
    shape_id = r['노선번호']
    print(str(index + 1) + '/' + str(len(R)) + ' (' + str((index+1)/len(R)*100) + '%)')
    headway = r['평균배차간격']    
    if (pd.isnull(headway)):
        continue
    
    for i in range(r['인가운행횟수']):
        trip_id = str(shape_id) + '_' + str(i + 1)
        arrival_time = pd.to_timedelta(r['기점첫차시간']) + (pd.to_timedelta(headway) * i)
        df_trips = df_trips.append(pd.DataFrame([[route_id, service_id, trip_id, shape_id]], columns=t_columns))

        for _, s in t.iterrows():
            arrival_time = timedelta_to_hhmmss(arrival_time + pd.to_timedelta(s['정류소 이동시간']))
            departure_time = timedelta_to_hhmmss(arrival_time + pd.to_timedelta(s['서비스시간']))
            stop_id = s['정류장ID']
            stop_sequence = s['정류소순번']
            df_stop_times = df_stop_times.append(pd.DataFrame([[trip_id, arrival_time, departure_time, stop_id, stop_sequence]], columns=s_columns))
        
    df_trips.to_csv('trips.txt', mode='a', index=False, encoding='utf-8-sig', header=False)
    df_stop_times.to_csv('stop_times.txt', mode='a', index=False, encoding='utf-8-sig', header=False)
```
